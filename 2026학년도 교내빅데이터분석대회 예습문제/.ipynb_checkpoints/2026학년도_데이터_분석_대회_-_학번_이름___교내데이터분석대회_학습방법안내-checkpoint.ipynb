## 2. 진짜 쓰는 함수는 5개뿐

| 하고싶은거 | 코드 |
|---|---|
| 그룹별 개수 | `df.groupby('col').size()` |
| 그룹별 합계 | `df.groupby('col').sum(numeric_only=True)` |
| 내림차순 정렬 | `df.sort_values('col', ascending=False)` |
| 인덱스 리셋 | `df.reset_index()` |
| 상위 N개 | `df.head(N)` |

이거 5개 조합으로 1~15번 다 풀림. 새로운 거 없음.

## 3. 실전에서 틀리는 이유 Top3

1. `encoding='cp949'` 안 씀 → 한글 깨짐
2. `sum()`할 때 문자열 컬럼 안 빼고 합산 시도 → 에러. `numeric_only=True` 쓰거나 `drop(columns=[...])`로 텍스트 컬럼 먼저 제거
3. `ascending=False` 안 붙여서 오름차순으로 나옴 (기본값이 오름차순이라 내림차순 원하면 꼭 붙여야 함)

## 4. 시험 직전 체크리스트

- [ ] `encoding='cp949'` 외웠나
- [ ] `groupby().sum(numeric_only=True)` 외웠나
- [ ] `sort_values(ascending=False)` 외웠나
- [ ] `reset_index()` 까먹지 않았나
- [ ] `plt.bar(x, y)` 손으로 한 번 써봤나

## 5. 공부 순서 (현실적으로 3일)

- **1일차**: crime.csv 열어서 구조 파악 + `groupby/sum` 손코딩 10번
- **2일차**: `sort_values + reset_index` + 컬럼명 바꾸기 손코딩 10번
- **3일차**: 시각화(`plt.bar`)까지 포함해서 1번부터 15번까지 안 보고 풀어보기
- **당일**: 치트시트(2번 표)만 옆에 두고 막히는 부분만 확인
