## 1. 필요 라이브러리 선언

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

## 2. 폰트 설정

```python
mpl.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

## 3. 데이터 불러오기  [ col 어떤 것이 있는지 확인하기 ]

```python
df = pd.read_csv('crime.csv', encoding='cp949')
df.head()
```

## 4. col에 어떤 것이 있는지 출력

```python
df.columns
```

## 5. '범죄대분류'만 다시 unique

```python
df['범죄대분류'].unique()
```

## 6. 범죄대분류 최대 빈도수 분석 [ 그룹화 : groupby ]

```python
count = df.groupby('범죄대분류').size()
count.sort_values(ascending=False)
```

## 7. 범죄대분류별 합계 sum

```python
total = df.groupby('범죄대분류').sum(numeric_only=True).sum(axis=1)
total
```

## 8. 범죄대분류별 합계 및 내림차순
### [ 인덱스(index) 재설정 ] [ col 이름 변경 ]

```python
result = total.sort_values(ascending=False).reset_index()
result.columns = ['category', 'total']
result
```

# 9. 최대 빈도수의 범죄대분류 리스트 생성

```python
namelist = result['category'].tolist()
namelist
```

## 10. 막대 그래프 시각화 [ 상위 5개 출력 ] 

```python
top5 = result.head(5)
plt.bar(top5['category'], top5['total'])
plt.title('범죄대분류별 발생건수 상위 5개')
plt.show()
```

## 11. '지능범죄' 중 범죄중분류 최대 빈도수 분석

```python
sub = df[df['범죄대분류'] == '지능범죄'].copy()
sub['total'] = sub.drop(columns=['범죄대분류', '범죄중분류']).sum(axis=1)
sub[['범죄중분류', 'total']].sort_values('total', ascending=False).reset_index(drop=True)
```
